##### 1. How do fare evasion rates vary across different hours of the day and days of the week, and what are the top three time periods with the highest fare evasion occurrences?

```python
top_fare_evasion = fare_evasion_df.nlargest(3, 'Fare_Evasion')
```

```python
daily_ridership_df["Date"] = pd.to_datetime(daily_ridership_df["Date"])
```

```python
daily_ridership_df["Day_of_Week"] = daily_ridership_df["Date"].dt.day_name()
```

```python
weekly_ridership = daily_ridership_df.groupby("Day_of_Week")["Count"].sum().reset_index() # sum of ridership by day of the week
```

```python
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekly_ridership["Day_of_Week"] = pd.Categorical(weekly_ridership["Day_of_Week"], categories=days_order, ordered=True)
weekly_ridership = weekly_ridership.sort_values("Day_of_Week")
```

```python
plt.figure(figsize=(7, 5))
plt.bar(weekly_ridership["Day_of_Week"], weekly_ridership["Count"], color="royalblue")
for i, v in enumerate(weekly_ridership["Count"]):
    plt.text(i, v, f"{v:,.0f}", ha="center", va="bottom", fontsize=7)

plt.xlabel("Day of the Week")
plt.ylabel("Estimated Fare Evasion (Based on Ridership)")
plt.title("Estimated Fare Evasion Across Different Days of the Week")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

##### 4.  Is there a correlation between fare evasion rates and the frequency of traffic violations, and do time periods with high fare evasion also have a higher number of traffic violations?


```python
# columns for fare evasion
columns = fare_evasion_df.columns
print(columns)
```

```python
traffic_violations_df["Quarter"] = ((traffic_violations_df["Violation_Month"] - 1) // 3 + 1).astype(str)
traffic_violations_df["Time_Period"] = traffic_violations_df["Violation_Year"].astype(str) + "-Q" + traffic_violations_df["Quarter"]
```

```python
# Aggregate traffic violations by Time_Period
traffic_violations_summary = traffic_violations_df.groupby("Time_Period").size().reset_index(name="Total_Traffic_Violations")
```

```python
# Ensure both datasets have Time_Period as string type
fare_evasion_df["Time_Period"] = fare_evasion_df["Time_Period"].astype(str)
traffic_violations_summary["Time_Period"] = traffic_violations_summary["Time_Period"].astype(str)
```

```python
merged_df = fare_evasion_df.merge(traffic_violations_summary, on="Time_Period", how="inner")
```

```python

correlation = merged_df["Fare_Evasion"].corr(merged_df["Total_Traffic_Violations"])

plt.figure(figsize=(8, 5))
plt.scatter(merged_df["Fare_Evasion"], merged_df["Total_Traffic_Violations"], color="royalblue", alpha=0.7)
plt.xlabel("Fare Evasion Rate")
plt.ylabel("Traffic Violations Count")
plt.title(f"Correlation Between Fare Evasion and Traffic Violations (r = {correlation:.2f})")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

```

##### 8. What are the most frequently occurring violations in subway stations, and which months experience the highest violation rates?

```python
top_violations = traffic_violations_df["Violation_Description"].value_counts().nlargest(10).reset_index()
top_violations.columns = ["Violation_Description", "Count"]
```

```python
monthly_violations = traffic_violations_df.groupby("Violation_Month").size().reset_index(name="Total_Violations")
```

```python

plt.figure(figsize=(10, 5))
bars = plt.barh(top_violations["Violation_Description"], top_violations["Count"], color="royalblue")

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"{bar.get_width():,.0f}", 
             ha="left", va="center", fontsize=7)

plt.xlabel("Number of Occurrences")
plt.ylabel("Violation Type")
plt.title("Top 10 Most Frequently Occurring Violations in Subway Stations")
plt.gca().invert_yaxis()  
plt.grid(axis="x", linestyle="--", alpha=0.7)

plt.show()

plt.figure(figsize=(10, 5))
bars = plt.bar(monthly_violations["Violation_Month"], monthly_violations["Total_Violations"], color="royalblue")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{bar.get_height():,.0f}", 
             ha="center", va="bottom", fontsize=7)

plt.xlabel("Month")
plt.ylabel("Total Violations")
plt.title("Violation Trends Across Different Months")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()


top_violations
```
